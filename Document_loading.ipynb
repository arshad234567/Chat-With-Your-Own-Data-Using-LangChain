## Youtube

```python
import whisper
import glob
import os

from langchain_core.documents import Document
from langchain_community.document_loaders.blob_loaders.youtube_audio import YoutubeAudioLoader

url = "https://www.youtube.com/watch?v=jGwO_UgTS7I"

save_dir = r"C:\Users\arsha\OneDrive\Desktop\Langchain_chat_with_your_own_data\docs\youtube"

# Download audio
YoutubeAudioLoader([url], save_dir).yield_blobs()

# Find audio file
audio_files = glob.glob(os.path.join(save_dir, "*.m4a"))

# Load local whisper model
model = whisper.load_model("base")

# Transcribe audio
result = model.transcribe(audio_files[0])

text = result["text"]

# Convert to LangChain document
docs = [
    Document(
        page_content=text,
        metadata={"source": url}
    )
]
```
